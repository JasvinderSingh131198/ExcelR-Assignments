# Assignment 7 (Clustering) Q2. Crime_Data (DBSCAN)

## Q2.  Perform Clustering(Hierarchical, Kmeans & DBSCAN) for the crime data and identify the number of clusters formed and draw inferences.

## Data Description:
## Murder -- Muder rates in different places of United States
## Assualt- Assualt rate in different places of United States
## UrbanPop - urban population in different places of United States
## Rape - Rape rate in different places of United States


```python
# Importing Required Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
# Importing the Dataset
crime_df=pd.read_csv('crime_data.csv')
crime_df
```

```python
crime_df.info()
```

```python
crime_df.isnull().sum()
```

```python
crime_df_corr = crime_df.corr()
print('Correlation : \n',crime_df_corr)
sns.heatmap(crime_df_corr,cmap='YlGnBu',annot=True)
```

```python
crime_df.drop(['Unnamed: 0'],axis=1,inplace=True)
crime_df
```

```python
# Normalizing Heterogenous Numerical Data using Standard Scalar Fit Transform to Dataset
crime_df_norm=StandardScaler().fit_transform(crime_df)
crime_df_norm
```

```python
# DBSCAN Clustering
dbscan=DBSCAN(eps=1,min_samples=4)
dbscan.fit(crime_df_norm)
```

```python
# Noisy samples are Given the Label -1.
dbscan.labels_
```

```python
# Adding Clusters to Dataset
crime_df['Clusters'] = dbscan.labels_
crime_df
```

```python
crime_df.groupby('Clusters').agg(['mean']).reset_index()
```

```python
# Plotting Clusters
plt.figure(figsize=(10, 7))  
plt.scatter(crime_df['Clusters'],crime_df['Murder'], c = dbscan.labels_)
```

```python
plt.figure(figsize=(10, 7))  
plt.scatter(crime_df['Clusters'],crime_df['Assault'], c = dbscan.labels_)
```

```python
plt.figure(figsize=(10, 7))  
plt.scatter(crime_df['Clusters'],crime_df['UrbanPop'], c = dbscan.labels_)
```

```python
plt.figure(figsize=(10, 7))  
plt.scatter(crime_df['Clusters'],crime_df['Rape'], c = dbscan.labels_)
```

```python
# KMeans Clustering
kmeans = KMeans(4)
kmeans.fit(crime_df)
```

```python
# Predictions
identified_clusters = kmeans.fit_predict(crime_df)
identified_clusters
```
