## Q.  Problem statement. Build a recommender system by using cosine simillarties score.

```python
# Importing Libraries
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Loading The Dataset
books_df = pd.read_csv('book.csv',encoding='latin1')
books_df
```

```python
books_df.info()
```

```python
books_df.isnull().sum()
```

```python
books_df1 = books_df.drop('Unnamed: 0',axis=1)
```

```python
books_df1
```

```python
books_df1.describe()
```

```python
# Implementing Count Vectorizer
cv = CountVectorizer()
count_matrix = cv.fit_transform(books_df1)
print("Count Matrix:", count_matrix.toarray())
```

```python
# Implementing cosine similarity
cosine_sim = cosine_similarity(count_matrix)
```

```python
cosine_sim
```

```python
# Giving Input Value for Book to check similar books
books_user_likes = "The Middle Stories"
def get_index_from_title(title):
    return books_df1[books_df1['Book.Title'] == books_user_likes]
books_index = get_index_from_title(books_user_likes)
```

```python
# Getting List of Cosine Similarities for books
similar_books = list(cosine_sim)
```

```python
# Sorting the Similar Books List in Descending Order
sorted_similar_books = sorted(similar_books, key=lambda x:x[1], reverse=True)
```

```python
# Printing Similar Books
def get_title_from_index(index):
    return books_df1[books_df1.index == index]["Book.Title"].values
i=0
for books in sorted_similar_books:
    print(get_title_from_index(books[0]))
    i=i+1
    if i>15:
        break
```
