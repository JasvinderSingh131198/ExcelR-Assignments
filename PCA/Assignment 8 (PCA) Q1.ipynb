```python
# Importing Required Libraries
import pandas as pd 
import numpy as np
import seaborn as sns
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
# Importing The Dataset
wine_df = pd.read_csv('wine.csv')
wine_df
```

```python
wine_df1 = wine_df.drop(['Type'],axis=1)
wine_df1
```

```python
wine_df1.isnull().sum()
```

```python
wine_df1.info()
```

```python
wine_df1.describe()
```

```python
sns.heatmap(wine_df1.corr(),cmap='YlGnBu',annot=True)
```

```python
# Normalizing The Data 
wine_norm = scale(wine_df1)
```

```python
wine_norm
```

```python
pca = PCA()
pca_values = pca.fit_transform(wine_norm)
```

```python
pca_values
```

```python
# The Amount of Variance that each PCA explains  
variance = pca.explained_variance_ratio_
variance
```

```python
# Cumulative Variance 
variance1 = np.cumsum(np.round(variance,decimals = 4)*100)
variance1
```

```python
pca.components_
```

```python
# Variance Plot for PCA components obtained 
plt.plot(variance1,color="red")
```

```python
pca_values[:,0:1]
```

```python
# plot between PCA1 and PCA2 
x = pca_values[:,0:1]
y = pca_values[:,9]
plt.scatter(x,y)
```

```python
sns.scatterplot(data=wine_df,x=wine_df['Alcohol'],y=wine_df['Color'],hue=wine_df['Type'])
```

```python
wine_df['Type'].value_counts()
```

```python
scaler = StandardScaler()
scaled_wine_df = scaler.fit_transform(wine_df.iloc[:,1:])
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(pca_values)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1, 11), wcss,marker='o',linestyle='--')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
clusters_new = KMeans(3, random_state=42)
clusters_new.fit(scaled_wine_df)
clusters_new.labels_
```

```python
clusters_new.fit(pca_values)
```

```python
wine_df_kmeans_pca = pd.concat([wine_df1.reset_index(drop=True),pd.DataFrame(pca_values)],axis=1)
wine_df_kmeans_pca.columns.values[-3:] = ['Type1','Type2','Type3']
wine_df_kmeans_pca['Wine Kmeans PCA'] = clusters_new.labels_
```

```python
wine_df_kmeans_pca
```

```python
wine_df_kmeans_pca['Types'] = wine_df_kmeans_pca['Wine Kmeans PCA'].map({0:'First',1:'Second',2:'Third'})
```

```python
plt.figure(figsize = (10,8))
sns.scatterplot(x=wine_df_kmeans_pca['Type2'],y=wine_df_kmeans_pca['Type1'],hue=wine_df_kmeans_pca['Types'])
plt.title('Clusters By PCA Components')
plt.show()
```
