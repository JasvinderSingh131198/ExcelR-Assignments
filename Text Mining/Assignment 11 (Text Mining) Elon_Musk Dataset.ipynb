```python
# Loading the Dataset
elon_musk_df = pd.read_csv('Elon_Musk.csv',encoding='latin1')
elon_musk_df
```

## Getting Word Count 

```python
elon_musk_df['Word_Count'] = elon_musk_df['Text'].apply(lambda x: len(str(x).split(" ")))
elon_musk_df
```

## Getting Character Count 

```python
elon_musk_df['Character_Count'] = elon_musk_df['Text'].str.len()
elon_musk_df
```

```python
elon_musk_df.info()
```

```python
elon_musk_df.isnull().sum()
```

```python
# Converting data into lowercase
elon_musk_df['Text'] = elon_musk_df['Text'].apply(lambda x:" ".join(x.lower() for x in x.split()))
elon_musk_df
```

```python
# Removing Punctuation
elon_musk_df['Text'] = elon_musk_df['Text'].str.replace('[^\w\s]','')
elon_musk_df
```

```python
import nltk
nltk.download('stopwords')
```

```python
# Removal of Stop Words
from nltk.corpus import stopwords
stop = stopwords.words('english')
elon_musk_df['Text'] = elon_musk_df['Text'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
elon_musk_df
```

```python
# Common Words Retrieval
freq = pd.Series(' '.join(elon_musk_df['Text']).split()).value_counts()
freq
```

```python
# Spelling Corrections
from textblob import TextBlob
elon_musk_df['Text'][:5].apply(lambda x: str(TextBlob(x).correct()))
```

```python
import nltk
nltk.download('punkt')
```

```python
# Tokenization
from nltk import tokenize
TextBlob(elon_musk_df['Text'][2]).words
```

```python
# Stemming
from nltk.stem import PorterStemmer
st = PorterStemmer()
elon_musk_df['Text'][:5].apply(lambda x: " ".join([st.stem(word) for word in x.split()]))
```

```python
import nltk
nltk.download('wordnet')
```

```python
# Lemmatization
from textblob import Word
elon_musk_df['Text'] = elon_musk_df['Text'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
elon_musk_df['Text'].head()
```

```python
# N-grams
TextBlob(elon_musk_df['Text'][0]).ngrams(3)
```

```python
# Bag of Words
from sklearn.feature_extraction.text import CountVectorizer
bag = CountVectorizer(max_features=1000,lowercase=True,ngram_range=(1,1),analyzer='word')
data_bag = bag.fit_transform(elon_musk_df['Text'])
data_bag
```
